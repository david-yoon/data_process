```python
import chardet

def find_encoding(filename):
    rawdata = open(filename, 'rb').read()
    result = chardet.detect(rawdata)
    charenc = result['encoding']    
    return charenc
```

```python
def my_encoder(my_string):
    for i in my_string:
        try :
            yield unicode(i)
        except UnicodeDecodeError:
            yield ' '
```

# merge CSV file

1. csv 형식 [ 이전화자, 현재화자, 이전 발화, 현재 발화, 이전정보 index, 현재정보 index ]
2. 연속된 서로 다른 화자의 대화만을 저장
3. scene 이 변경 되면 이어지지 않는 대화로 간주

```python
filenames = os.listdir('friends/02_raw_csv/')
```

```python

f_out = codecs.open('./friends/csv_all.csv', 'w', encoding='utf-8')
writer = csv.writer(f_out, delimiter=',') 

for f_name in filenames:
    
    print "process : " + f_name
    
    target =  './friends/02_raw_csv/'+ f_name
    enc = find_encoding(target)
    
    with codecs.open(target, 'r', encoding=enc ) as f:
        
        prev = ""
        cur = ""
        
        reader = csv.reader(f, delimiter=',')
        for row in reader:
            
            cur = row

            if cur[0].lower() == 'scene':
                prev = ""
                continue
            
            if prev == "" :
                prev = cur
                continue
            
            if cur[0].strip() == prev[0].strip() :
                # 화자가 변하지 않은 경우
                continue
            else:
                # 화자가 변한 경우 -> 대화 저장
                writer.writerow( [ prev[0].strip(), 
                                          cur[0].strip(), 
                                          prev[1].strip(), 
                                          cur[1].strip(), 
                                          prev[2].strip(), 
                                          cur[2].strip() ]
                                       )
            prev = cur
    
f_out.close()        
```

# divide file

```python
import numpy as np
```

```python
def divide_file( filename, prob):
    
    target = './friends/03_csv/' + filename
    enc = find_encoding(target)
    
    f_read = codecs.open( target, 'r', encoding=enc )
    csv_reader = csv.reader(f_read, delimiter=',')

    f_out1 = codecs.open( './friends/03_csv/' + filename + '_part1', 'w', encoding='utf-8' )
    writer1 = csv.writer(f_out1, delimiter=',')

    f_out2 = codecs.open( './friends/03_csv/' + filename + '_part2', 'w', encoding='utf-8' )
    writer2 = csv.writer(f_out2, delimiter=',')

    for row in csv_reader:
        if prob > np.random.rand(1)[0]:
            writer1.writerow( row )
        else:
            writer2.writerow( row )

    f_read.close()
    f_out1.close()
    f_out2.close()
```

```python
filename = 'csv_all.csv'
prob = 0.15

divide_file(filename, prob)
```

# simple merge

```python
import csv 

def merge_file( filename1, filename2):
    f_read1 = open( './friends/03_csv/' + filename1, 'rb' )
    csv_reader1 = csv.reader(f_read1, delimiter=',')
    
    f_read2 = open( './friends/03_csv/' + filename2, 'rb' )
    csv_reader2 = csv.reader(f_read2, delimiter=',')

    f_out = open( './friends/03_csv/' + filename1.split('.')[0] + '_' + filename2, 'wb' )
    writer = csv.writer(f_out, delimiter=',')

    for row in csv_reader1:
        writer.writerow( row )
        
    for row in csv_reader2:
        writer.writerow( row )
        
    f_read1.close()
    f_read2.close()
    f_out.close()
```
